```python
import pickle
from tqdm.auto import tqdm
with open('sentences_segmented.pkl', 'rb') as handle:
        period_sentences_dict = pickle.load(handle)
        print(f"Loaded data for {len(period_sentences_dict)} periods")
```

```python
target_words = ["心里", "心中", "心上", "心里头", "心头",
                     "心下", "心内", "满心", "内心", "胸中", 
                     "心底", "心口", "心间", "心低", "心肝", 
                     "肚里", "腹内", "胸口", "心坎", "心房",
                     "脑子里"]
replaced = 0 
for period, sentences in tqdm(period_sentences_dict.items()):
    for sentence in tqdm(sentences):
        for word_idx,word in enumerate(sentence):
            if word in target_words:
                 sentence[word_idx] = "heart" # replace the word with "heart"
                 replaced += 1

print(f"Replaced {replaced} words")
```

```python
window = 20
for period_id, (period, sentences) in enumerate(period_sentences_dict.items()):
    with open(f"heart_{period_id}.txt", 'w') as f:
        print(period)
        for sentence_id, sentence in enumerate(sentences):
            for word_idx, word in enumerate(sentence):
                if word == "心上":
                    start_index = max(0, word_idx - window)
                    end_index = min(len(sentence), word_idx + window)
                    fragment = sentence[start_index:end_index]
                    print(f"{period_id}:{sentence_id}: {"".join(sentence)}", file=f)
```
